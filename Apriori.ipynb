```python
import pandas as pd
from mlxtend.frequent_patterns import apriori, association_rules
```

```python
dataset = [
    ['Milk', 'Bread'],
    ['Milk', 'Diaper', 'Beer', 'Eggs'],
    ['Milk', 'Diaper', 'Beer', 'Coke'],
    ['Bread', 'Milk', 'Diaper', 'Beer'],
    ['Bread', 'Milk', 'Diaper', 'Coke']
]
```

```python
from mlxtend.preprocessing import TransactionEncoder

te = TransactionEncoder()
te_data = te.fit(dataset).transform(dataset)

df = pd.DataFrame(te_data, columns=te.columns_)
print(df)
```

```python
frequent_items = apriori(df, min_support=0.6, use_colnames=True)
print(frequent_items)
```

```python
rules = association_rules(frequent_items, metric="confidence", min_threshold=0.7)
print(rules)
```

```python
import matplotlib.pyplot as plt

# Sort by support
frequent_items.sort_values(by='support', ascending=False, inplace=True)

# Convert itemsets to string for plotting
frequent_items['itemsets'] = frequent_items['itemsets'].astype(str)

# Plot
plt.figure()
plt.bar(frequent_items['itemsets'], frequent_items['support'])
plt.xticks(rotation=45)
plt.xlabel("Itemsets")
plt.ylabel("Support")
plt.title("Frequent Itemsets vs Support")
plt.show()
```

```python
import networkx as nx

G = nx.DiGraph()

for _, row in rules.iterrows():
    G.add_edge(str(row['antecedents']), str(row['consequents']),
               weight=row['lift'])

plt.figure()
pos = nx.spring_layout(G)
nx.draw(G, pos, with_labels=True)
nx.draw_networkx_edge_labels(G, pos)
plt.title("Association Rules Network")
plt.show()
```
