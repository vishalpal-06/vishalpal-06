```python
students = {"101":"Amit","102":"Neha","103":"Raj"}
students
```

```python
marks = {"101":85,"102":90,"103":78}
marks
```

```python
join_result = [(sid, students[sid], marks[sid]) for sid in students]
join_result
```

```python
selected = [(sid, name, marks[sid]) for sid,name in students.items() if marks[sid] > 80]
selected
```

```python
projection = [(name, marks[sid]) for sid,name in students.items()]
projection
```
