```python
import nltk
import pandas as pd
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer
```

```python
documents = [
    "Data Science is an amazing field.",
    "Machine learning makes predictions easier.",
    "Text mining extracts useful information.",
    "Python is widely used for data analysis.",
    "NLP helps computers understand human language."
]
```

```python
documents = [doc.lower() for doc in documents]
print(documents)
```

```python
nltk.download('punkt')
nltk.download('punkt_tab') # Download punkt_tab as suggested by the error message
tokenized_docs = [word_tokenize(doc) for doc in documents]
print(tokenized_docs)
```

```python
nltk.download('stopwords')
stop_words = set(stopwords.words('english'))
filtered_docs = [[word for word in doc if word not in stop_words] for doc in tokenized_docs]
print(filtered_docs)
```

```python
stemmer = PorterStemmer()
stemmed_docs = [[stemmer.stem(word) for word in doc] for doc in filtered_docs]
print(stemmed_docs)
```

```python
processed_docs = [" ".join(doc) for doc in stemmed_docs]
print(processed_docs)
```

```python
cv = CountVectorizer()
bow = cv.fit_transform(processed_docs)
print(pd.DataFrame(bow.toarray(), columns=cv.get_feature_names_out()))
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(processed_docs)
print(pd.DataFrame(tfidf_matrix.toarray(), columns=tfidf.get_feature_names_out()))
```
