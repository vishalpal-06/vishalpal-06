```python
import os, shutil

# Create a simulated HDFS root in local system
HDFS_ROOT = "/content/hdfs"
os.makedirs(HDFS_ROOT, exist_ok=True)

def hdfs_mkdir(path):
    os.makedirs(HDFS_ROOT + path, exist_ok=True)
    print("Directory created:", HDFS_ROOT + path)

def hdfs_put(local, hdfs):
    shutil.copy(local, HDFS_ROOT + hdfs)
    print("Uploaded:", local, "→", HDFS_ROOT + hdfs)

def hdfs_cat(hdfs):
    with open(HDFS_ROOT + hdfs) as f:
        print(f.read())

# Test commands
hdfs_mkdir("/employee")
with open("/content/sample.txt", "w") as f:
    f.write("This is a test employee document for test employee")

hdfs_put("/content/sample.txt", "/employee/sample.txt")
hdfs_cat("/employee/sample.txt")
```

```python
from collections import defaultdict

def mapper(line):
    words = line.strip().split()
    return [(w, 1) for w in words]

def reducer(mapped_pairs):
    result = defaultdict(int)
    for word, count in mapped_pairs:
        result[word] += count
    return dict(result)


# Run WordCount MapReduce
lines = open("/content/sample.txt").read().split("\n")

mapped = []
for line in lines:
    mapped.extend(mapper(line))

reduced = reducer(mapped)
reduced
```
